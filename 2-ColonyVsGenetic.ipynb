```python
import numpy as np
import random
import time
from matplotlib import pyplot as plt

# ant colony algorythm
from Tools.colony import ant_colony
# genetic algorythm
from Tools.genetic import genetic_algorythm
```

# Running Algorythms For Diffrent Number Of Cities

```python
# Inicializing arrays that will store results
times = np.empty((0, 2))
distances = np.empty((0, 2))
n_cities = np.array([])
```

```python
for n in range(10, 60, 5):
    
    # Inicializing cities
    start=np.array([random.random(),random.random()]) # startowe
    cities=np.array([[random.random(),random.random()] for _ in range(n-1)]) # po≈õrednie
    
    # genetic alg
    tourmin, distance, time  = genetic_algorythm(start, cities)
    # ant colony alg
    best_route, min_l, exec_time, fero = ant_colony(cities_=np.vstack([start, cities]), iter_=500)
    
    # Updating arrays
    times = np.append(times, [[time, exec_time]], axis=0)
    distances = np.append(distances, [[distance, min_l]], axis=0)
    n_cities = np.append(n_cities, n)
```

```python
print(times.shape)
print(distances.shape)
print(n_cities.shape)
```

# Visualization

```python
# Time Comparison Plot
plt.figure(figsize=(8,5))
plt.title("Time Comparison", fontsize=16) 

plt.scatter(n_cities, times[:,0], c="blue", label="Genetic Alg")
plt.scatter(n_cities, times[:,1], c="red", label="Ant Colony")

plt.grid()
plt.legend()
plt.xlabel('number of cities',fontsize=18);
plt.ylabel('time of execution',fontsize=18);
```

```python
# Distances Comparison Plot
plt.figure(figsize=(8,5))
plt.title("Distance Comparison", fontsize=16) 

plt.scatter(n_cities, distances[:,0], c="blue", label="Genetic Alg")
plt.scatter(n_cities, distances[:,1], c="red", label="Ant Colony")

plt.grid()
plt.legend()
plt.xlabel('number of cities',fontsize=18);
plt.ylabel('distance',fontsize=18);
```
